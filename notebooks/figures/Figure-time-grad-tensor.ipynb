```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('../stats_grads_tensor_experimental_101025.csv', index_col=0)
```

```python
sns.lineplot(df.query('iterations > 0'), x='iterations', y='time')
plt.xlabel('veros_step')
plt.ylabel('time (seconds)')
plt.title('Backward compute time (Quadro P6000 - ACC 1°)')

plt.savefig(f'backward_time_tensor_{datetime.now().strftime("%d%m%y")}.png')
```
